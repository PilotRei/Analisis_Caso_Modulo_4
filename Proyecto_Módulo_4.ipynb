##LECCIÓN 2: CONCEPTOS BÁSICOS DE ESTADÍSTICA DESCRIPTIVA


1.   Objetivo:

  Aplicar conceptos de tendencia central, dispersión y
medidas de posición.
  
2.   Tareas a desarrollar:

*   Calcular media, mediana, moda, varianza, desviación
estándar
*   Determinar cuartiles y percentiles
*   Generar histogramas y boxplots
*   Identificar valores atípicos y documentar su impacto

```python
import matplotlib.pyplot as plt


print("\n\n ----------------------LECCIÓN 2: ESTADÍSTICA DESCRIPTIVA ------------------------\n\n")

print("=============================================================================")
print("Tarea_1")
print("Cálculo de Medidas de Tendencia Central y Dispersión")
print("=============================================================================")


media_gasto = df['Monto_Gasto'].mean() #Media de los Gastos
mediana_gasto = df['Monto_Gasto'].median() #Mediana de los Gastos
moda_suscripcion = df['Suscripcion'].mode() #Moda de la Suscripción
varianza_gasto = df['Monto_Gasto'].var() # Varianza de los Gastos
desviacion = df['Monto_Gasto'].std() # Desv. Est. de los Gastos

print(f"Media del Gasto: $ {media_gasto:.2f}")
print(f"Mediana del Gasto: $ {mediana_gasto:.2f}")
print(f"Moda de Suscripción: {moda_suscripcion}")
print(f"Varianza del Gasto: $² {varianza_gasto:.2f}")
print(f"Desviación Estándar: $ {desviacion:.2f}")

# Resumen rápido con Pandas
print("\n\n--------------------- Resumen Estadístico Completo ---------------------\n")
print(df.describe())

print("\n\n")
print("=============================================================================")
print("Tarea_2")
print("MEDIDAS DE POSICIÓN (Cuartiles y Percentiles)")
print("=============================================================================")


# Cuartiles (Q1, Q2, Q3)
# Q1 = 25%, Q2 = 50% (Mediana), Q3 = 75%
cuartiles = df['Monto_Gasto'].quantile([0.25, 0.5, 0.75])

# Percentil específico (ejemplo: Percentil 90)

percentil_90 = df['Monto_Gasto'].quantile(0.90)

print("\n\n--- Cuartiles de Monto de Gasto ---\n")

print(f"Cuartil 1 (25%): ${cuartiles[0.25]:.2f}")
print(f"Cuartil 2 (50% - Mediana): ${cuartiles[0.50]:.2f}")
print(f"Cuartil 3 (75%): ${cuartiles[0.75]:.2f}")

print(f"\nPercentil 90: ${percentil_90:.2f}")
print("         ( El 90% de los clientes gasta un monto igual o menor a este valor )")

print("\n\n")
print("=============================================================================")
print("Tarea_3")
print("HISTOGRAMA y BOXPLOT")
print("=============================================================================")

# Configuramos el área de dibujo para mostrar dos gráficos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# 1. HISTOGRAMA: Para ver la distribución (Frecuencia de montos)
ax1.hist(df['Monto_Gasto'].dropna(), bins=10, color='skyblue', edgecolor='black')
ax1.set_title('Distribución de Montos de Gasto')
ax1.set_xlabel('Monto ($)')
ax1.set_ylabel('Cantidad de Clientes')

# 2. BOXPLOT: Para una mejor detección de Outliers (Valores atípicos)
ax2.boxplot(df['Monto_Gasto'].dropna(), vert=False)
ax2.set_title('Detección de Valores Atípicos (Outliers)')
ax2.set_xlabel('Monto ($)')

plt.tight_layout()
plt.show()
```

```python

print("\n\n ----------------------LECCIÓN 3: CORRELACIÓN ------------------------\n\n")

print("=============================================================================")
print("Tarea_1")
print("Matriz de correlación (Pearson)")
print("=============================================================================")

# Filtramos solo las columnas numéricas para evitar errores
columnas_numericas = df.select_dtypes(include=['number'])
matriz_corr = columnas_numericas.corr()

print(matriz_corr)

# Extraemos la relación específica entre Visitas y Gasto
r_visitas_gasto = matriz_corr.loc['Visitas_Mensuales', 'Monto_Gasto']
print(f"\n\n Coeficiente R entre Visitas y Gasto: {r_visitas_gasto:.2f}")

print("\n\n")
print("=============================================================================")
print("Tarea_2")
print("Scatterplot: Visitas vs Gasto")
print("=============================================================================")


plt.figure(figsize=(8, 6))
plt.scatter(df['Visitas_Mensuales'], df['Monto_Gasto'], color='purple', alpha=0.6)

# Añadimos etiquetas para que sea un informe profesional
plt.title('Relación entre Visitas Mensuales y Monto de Gasto')
plt.xlabel('Número de Visitas')
plt.ylabel('Monto Gasto ($)')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
